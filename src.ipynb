```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
placement_per_day = np.random.poisson(lam = 4,size = 4200)
print(placement_per_day)
```

```python
x_axis = np.unique(placement_per_day)
```

```python
for x in x_axis:
    frequency = np.count_nonzero(placement_per_day == x)
    relative_frequency = frequency/placement_per_day.shape[0]
    relative_frequency
    print(relative_frequency)
```

```python
weights = np.random.normal(loc=65,scale=5,size=10000000)
z_score = (weights - np.mean(weights))/np.std(weights)
axes = plt.hist(z_score,bins=500)
```
